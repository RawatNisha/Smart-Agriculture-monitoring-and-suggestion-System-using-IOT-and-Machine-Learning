## Training Paths

## Testing Paths

```python
Testing_Paths=[
    "../../Data/Main_Data/Testing_Data/Rice/Bacterial leaf blight",
    "../../Data/Main_Data/Testing_Data/Rice/Brown spot",
    "../../Data/Main_Data/Testing_Data/Rice/Leaf smut",
    "../../Data/Main_Data/Testing_Data/Rice/Normal",
]
```

```python
os.listdir(Training_Paths[0])
```

```python
class_list=[
    "Bacterial leaf blight",
    "Brown spot",
    "Leaf smut",
    "Normal"
]
```

```python
Training_Data=[]
counter=0
index=0
for class_ in Training_Paths:
    print(class_)
    for file in os.listdir(class_):
        print(file)
        counter+=1;
        image_array=cv2.imread(os.path.join(class_,file))
        print(image_array)
        image_array=cv2.cvtColor(image_array,cv2.COLOR_BGR2RGB)
        image_array=cv2.resize(image_array,(160,160))
        Training_Data.append([image_array/255,index])
#         pyplot.imshow(image_array,cmap='gray')
    index+=1
print(counter)
```

```python
Training_Data
```

```python
random.shuffle(Training_Data)
```

```python
Training_Data
```

```python
Training_features=[]
Training_Labels=[]
for value in Training_Data:
    Training_features.append(value[0])
    Training_Labels.append(value[1])
```

```python
Training_features
```

```python
Training_Labels
```

```python
pickle_Training_features=open("../Data/Data_Object/Training_Objects/Pickle_Training_features.pickle","wb")
pickle.dump(Training_features,pickle_Training_features)
pickle_Training_features.close()
pickle_Training_labels=open("../Data/Data_Object/Training_Objects/Pickle_Training_labels.pickle","wb")
pickle.dump(Training_Labels,pickle_Training_labels)
pickle_Training_labels.close()
```

```python
Testing_Data=[]
```

```python
counter=0
index=0
for class_ in Testing_Paths:
    print(class_)
    print(len(os.listdir(class_)))
    for file in os.listdir(class_):
        print(file)
        counter+=1;
        image_array=cv2.imread(os.path.join(class_,file))
        image_array=cv2.cvtColor(image_array,cv2.COLOR_BGR2RGB)
        image_array=cv2.resize(image_array,(160,160))
        Testing_Data.append([image_array/255,index])
#         pyplot.imshow(image_array,cmap='gray')
    index+=1
print(counter)
random.shuffle(Testing_Data)
Testing_features=[]
Testing_Labels=[]
for value in Testing_Data:
    Testing_features.append(value[0])
    Testing_Labels.append(value[1])
```

```python
pickle_Testing_features=open("../Data/Data_Object/Testing_Objects/Pickle_Testing_features.pickle","wb")
pickle.dump(Testing_features,pickle_Testing_features)
pickle_Testing_features.close()
pickle_Testing_labels=open("../Data/Data_Object/Testing_Objects/Pickle_Testing_labels.pickle","wb")
pickle.dump(Testing_Labels,pickle_Testing_labels)
pickle_Testing_labels.close()
```
